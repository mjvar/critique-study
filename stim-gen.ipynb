```python
import revisitpy as rvt
```

```python
pre_response = rvt.response(
    id='pre-response',
    prompt="In 3-5 words, how would you describe the visualization you just read?",
    type='shortText',
    location='belowStimulus'
)

post_change = rvt.response(
    id='post-change',
    prompt="Would you say that your opinion of the visualization has changed significantly after speaking about it?",
    type='radio',
    options=['Yes, I have changed my opinion', 'No, my opinion has not changed'],
    location='belowStimulus'
)

post_response = rvt.response(
    id='post-response',
    required=False,
    prompt="If your opinion has changed, please write how you would now describe the visualization in 3-5 words.",
    type='shortText',
    location='belowStimulus',
    secondaryText="If your opinion has not changed, you may leave this field blank.",
    placeholder="Enter your updated description here"
)

stimulus_read = rvt.component(
    component_name__='stimulus-read',
    type='image',
    path="",
    response=[],
    recordAudio=False,
    style={
        "width": "800px",
        "height": "auto"
    },
    instruction="Please read and understand the following visualization. Then, click 'Next' to proceed.",
    nextButtonEnableTime=30000,
    nextButtonText="Next"
)

stimulus_pre = rvt.component(
    component_name__='stimulus-pre',
    type='image',
    path="",
    recordAudio=False,
    style={
        "width": "800px",
        "height": "auto"
    },
    instruction="In 3-5 words, how would you describe the visualization you just read?",
    response=[pre_response]
)

stimulus_getready = rvt.component(
    component_name__='stimulus-getready',
    type='image',
    path="",
    recordAudio=False,
    style={
        "width": "800px",
        "height": "auto"
    },
    instruction="Please prepare to critique the visualization you just read. Your critiques may focus on strengths or weaknesses of the visualization. **When you are ready, click \"Start Recording\".**",
    nextButtonText="Start Recording",
    response=[]
)

stimulus_speak = rvt.component(
    component_name__='stimulus-speak',
    type='image',
    path="",
    recordAudio=True,
    style={
        "width": "800px",
        "height": "auto"
    },
    instruction="Please critique the visualization you just read. Your critiques may focus on strengths or weaknesses of the visualization. **Once you are finished speaking, click \"Finish Recording\".**",
    nextButtonText="Finish Recording",
    response=[]
)

stimulus_post = rvt.component(
    component_name__='stimulus-post',
    type='react-component',
    recordAudio=False,
    path="!pilot-thinkaloud/assets/StimulusPostRecall.tsx",
    style={
        "width": "800px",
        "height": "auto"
    },
    response=[post_change, post_response]
)

base_components = [stimulus_read, stimulus_pre, stimulus_getready, stimulus_speak, stimulus_post]
```
